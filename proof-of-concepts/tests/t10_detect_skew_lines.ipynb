```python
def rolling_percentile_2d(arr, sz, q):
    res = np.zeros((sz[0]*sz[1], arr.shape[0] - sz[0] + 1, arr.shape[1] - sz[1] + 1))
    
    for i in range(sz[0]):
        for j in range(sz[1]):
            res[i*sz[1]+j, :, :] = arr[i:(arr.shape[0]-sz[0]+1+i),j:(arr.shape[1]-sz[1]+1+j)]
    
    return np.percentile(res, q, axis=0)

def calc_variance_percentile_v(src, filt_sz, q):
    var = np.cumsum(np.abs(src[1:,:]-src[:-1,:]), axis=0)
    mean = (var[filt_sz[0]:,:]-var[:-filt_sz[0],:]) / filt_sz[0]
    return rolling_percentile_2d(mean, (1, filt_sz[1]), q)

def calc_variance_percentile_h(src, filt_sz, q):
    var = np.cumsum(np.abs(src[:,1:]-src[:,:-1]), axis=1)
    mean = (var[:,filt_sz[1]:]-var[:,:-filt_sz[1]]) / filt_sz[1]
    return rolling_percentile_2d(mean, (filt_sz[0], 1), q)

def local_maxima_filter_v(src, i, d1, d2, cutoff):
    ilo1 = max(0, i-d1)
    ilo2 = max(0, i-d2)
    ihi1 = min(src.shape[0], i+d1+1)
    ihi2 = min(src.shape[0], i+d2+1)
    m1 = np.max(src[ilo1:ihi1], axis=0)
    m2 = np.max(src[ilo2:ihi2], axis=0)
    return (m1 >= m2)*(m2 > cutoff)
```

```python
def longest_sequence(bool_array_1d, slack = 0):
    max_seq = 0
    max_ind = 0
    cur_seq = 0
    cur_slack = slack
    for i in range(len(bool_array_1d)):
        if bool_array_1d[i]:
            cur_slack = slack
            cur_seq += 1
        else:
            if cur_slack > 0:
                cur_slack -= 1
                cur_seq += 1
            else:
                if max_seq < cur_seq:
                    max_seq = cur_seq
                    max_ind = i
                cur_seq = 0
                cur_slack = slack
    if max_seq < cur_seq:
        max_seq = cur_seq
        max_ind = len(bool_array_1d)
    return (max_ind - max_seq, max_ind)

def closest_to_point_v(src, p, max_dist):
    ind_n = np.array(range(max(0, p[0]-max_dist), p[0]+1))[::-1]
    ind_p = np.array(range(p[0], min(src.shape[0], p[0]+max_dist+1)))
    pos_n = np.argmax(np.append(src[ind_n, p[1]]>0, True)) 
    pos_p = np.argmax(np.append(src[ind_p, p[1]]>0, True))
    if pos_n == max_dist + 1:
        pos_n = np.inf
    if pos_p == max_dist + 1:
        pos_p = np.inf
    return (-pos_n, pos_p)
    
def closest_to_line(src, p0, p1, max_dist=20):
    j0 = min(p0[1], p1[1])
    j1 = max(p0[1], p1[1])
    a = [(j-p0[1])/(p1[1]-p0[1]) for j in range(j0, j1+1)]
    p = [(int(np.round((1-a[i])*p0[0]+a[i]*p1[0])),int((1-a[i])*p0[1]+a[i]*p1[1])) for i in range(j1-j0+1)]
    return [closest_to_point_v(src, p[i], max_dist) for i in range(j1-j0+1)]

def calc_loss(dsts, dd, dist_cutoff, reg_coef, reg_power):
    num = len(dsts)-1
    tmp0 = [(d[0] if abs(d[0]) < dist_cutoff else np.inf, d[1] if abs(d[1]) < dist_cutoff else np.inf) for d in dsts]
    tmp1 = [min(abs(round(dd*i/num)-ds[0]), abs(round(dd*i/num)-ds[1])) for (i, ds) in enumerate(tmp0)]
    tmp2 = [x for x in tmp1 if not np.isinf(x)]
    if len(tmp2) == 0:
        return np.inf
    return np.sum(np.power(tmp2, 2.0))/num + reg_coef*np.power(abs(dd), reg_power)

def find_best_diff(dsts, max_dd, dist_cutoff, reg_coef, reg_power):
    r0 = calc_loss(dsts, 0, dist_cutoff, reg_coef, reg_power)
    l1 = [calc_loss(dsts, i, dist_cutoff, reg_coef, reg_power) for i in range(1, max_dd+1)]
    l2 = [calc_loss(dsts, -i, dist_cutoff, reg_coef, reg_power) for i in range(1, max_dd+1)]
    i1 = np.argmin(l1)
    i2 = np.argmin(l2)
    #print(l1)
    #print(l2)
    if r0 <= min(l1[i1], l2[i2]):
        return (0, r0)
    if l1[i1] <= min(r0, l2[i2]):
        return (i1+1, l1[i1])
    if l2[i2] <= min(r0, l1[i1]):
        return (-(i2+1), l2[i2])
    return None

def extend_line(src, p, p_new, dy, max_dist, dist_cutoff, reg_coef, reg_power):
    lt = closest_to_line(src, p, p_new, max_dist)
    (new_dy, loss) = find_best_diff(lt, dy, dist_cutoff, reg_coef, reg_power)
    return ((p_new[0]+new_dy, p_new[1]), loss)

def extend_multi_line(src, p, dx, dy, max_dist, dist_cutoff, pct_cutoff, reg_coef, reg_power):
    points = [p]
    p0 = p
    dy_prev = 0
    while True:
        x_new = min(src.shape[1]-1, max(0, p0[1]+dx))
        if x_new == p0[1]:
            break
        dy_act = int(np.ceil(abs(x_new - p0[1])/abs(dx)*dy))
        y_new = p0[0]+int(np.round((x_new - p0[1])/dx*dy_prev))
        if y_new >= src.shape[0] or y_new < 0:
            break
        (p1, loss) = extend_line(src, p0, (y_new, x_new), dy_act, max_dist, dist_cutoff, reg_coef, reg_power)
        
        dsts = closest_to_line(src, p0, p1, max_dist)
        adsts = [min(abs(x[0]), abs(x[1])) for x in dsts]
        num0 = len(adsts)
        num1 = len([x for x in adsts if x < dist_cutoff])
        
        if np.isinf(loss) or num1/num0 < pct_cutoff:
            break
        
            
        dy_prev = p1[0] - p0[0]
        p0 = p1
        points.append(p1)
    return points

def find_line(src, y0, dx, dy, max_dist, dist_cutoff, pct_cutoff, slack, reg_coef, reg_power):
    (x00,x01) = longest_sequence(src[y0, :]==1, slack)
    x0 = int(np.round(0.5*(x00+x01)))
    print(y0, x0)
    p1 = extend_multi_line(src, (y0, x0), -dx, dy, max_dist, dist_cutoff, pct_cutoff, reg_coef, reg_power)[::-1]
    p2 = extend_multi_line(src, (y0, x0), dx, dy, max_dist, dist_cutoff, pct_cutoff, reg_coef, reg_power)[1:]
    return p1+p2

def erase_line(src, lines, dy):
    res = np.copy(src)
    for (p1, p2) in zip(lines[:-1], lines[1:]):
        for j in range(p1[1], p2[1]+1):
            a = (j-p1[1])/(p2[1]-p1[1])
            i0 = int(np.round((1-a)*p1[0]+a*p2[0]))
            i1 = max(0, i0-dy)
            i2 = min(src.shape[0], i0+dy+1)
            res[i1:i2, j] = 0
    return res
```
