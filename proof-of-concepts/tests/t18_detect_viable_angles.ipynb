```python
def find_num_zero_regions(arr1d):
    return np.sum((arr1d[1:]<=0)&(arr1d[:-1]>0)) - 1

def find_angle_ids_range(arr2d):
    num_zeros = np.array([find_num_zero_regions(x) for x in arr2d])
    maxv = np.max(num_zeros)
    return np.arange(arr2d.shape[0])[num_zeros >= maxv]

def find_longest_consecutive_seq(seq):
    i0 = 0
    v0 = 0
    v_prev = -10
    v_len = 0
    for i in range(len(seq)):
        if v_prev+1 == seq[i]:
            v_prev = seq[i]
            v_len += 1
            continue
        v_prev = seq[i]
        if v_len > v0:
            v0 = v_len
            i0 = i - v0
        v_len = 1
    if v_len > v0:
        v0 = v_len
        i0 = len(seq) - v0
    return seq[i0:(i0+v0)]

def find_largest_angle_range(arr2d):
    rng = find_angle_ids_range(arr2d)
    #print(rng)
    res = find_longest_consecutive_seq(rng)
    return np.min(res), np.max(res)

def calc_width(arr1d, j, cutoff):
    n_l = np.argmin(arr1d[j:]<cutoff)
    n_r = np.argmin(arr1d[j::-1]<cutoff)
    if n_l == 0:
        n_l = len(arr1d)-j
    if n_r == 0:
        n_r = j        
    return min(n_r, n_l)

def find_split_width(arr1d, q=25, pct_split=1):
    msk = arr1d > 0
    if not np.any(msk):
        return [] #any offset suits basically    
    s0 = np.argmax(msk)
    s1 = len(arr1d) - np.argmax(msk[::-1])
    tmp1d = arr1d[s0:s1]

    cutoff = np.percentile(tmp1d, q)
    candidates = np.arange(len(tmp1d))[tmp1d<=0]
    res = [(s0+x, calc_width(tmp1d, x, cutoff), np.sum(tmp1d[:x]), np.sum(tmp1d[(x+1):])) for x in candidates]
    return[(ind, wid, min(a,b) / (a+b)) for (ind, wid, a, b) in res if min(a,b)/(a+b) > pct_split/100.0]

def find_best_split(arr2d, q=25, pct_split=1):
    (a1, a2) = find_largest_angle_range(arr2d)
    angle_id = (a1 + a2) // 2
    splits = find_split_width(arr2d[angle_id], q, pct_split)
    if len(splits)==0:
        return None
    max_width = np.max([x for (_,x,_) in splits])
    splits1 = [(x, w, pct) for (x,w,pct) in splits if w >= max_width]
    res = sorted(splits, key=lambda x: x[2], reverse=True)[0]
    return (angle_id, res[0], res[1], res[2])

def apply_split(src, radon_grid, angle_id, offset_id, i0, i1, j0, j1):
    res1 = np.copy(src)
    res2 = np.copy(src)
    goff = local_radon.global_offset_from_subgrid(radon_grid, i0, i1, j0, j1, angle_id, offset_id)
    res1[radon_grid[:,:,angle_id]<goff] = 0
    res2[radon_grid[:,:,angle_id]>goff] = 0
    return (res1, res2)
```

```python
def find_all_interior_zeros(arr2d):
    s0 = np.argmax(np.hstack((arr2d>0, np.ones((arr2d.shape[0],1))>0)), axis=1)
    s1 = arr2d.shape[1] - 1 - np.argmax(np.hstack((arr2d[:,::-1]>0, np.ones((arr2d.shape[0],1))>0)), axis=1)
    res0 = [(i, s0[i]+np.arange(s1[i]-s0[i]+1)[arr2d[i, s0[i]:(s1[i]+1)]<=0] if s1[i]>s0[i] else np.array([])) for i in range(len(s0))]
    return np.array([(i,j) for (i, arr) in res0 for j in arr])

def calc_line_stats(lns, angles, grid, i0, i1, j0, j1, t0, t1):
    res = []
    for (angle_id, offset_id) in lns:
        offset = local_radon.global_offset_from_subgrid(grid, i0, i1, j0, j1, angle_id, offset_id)
        angle = angles[angle_id]
        j0i = np.round((offset - t0*np.sin(angle))/np.cos(angle))
        j1i = np.round((offset - t1*np.sin(angle))/np.cos(angle))
        if j0i < i0:
            continue
        if j0i >= i1:
            continue
        if j1i < i0:
            continue
        if j1i >= i1:
            continue
        res.append((angle_id, offset_id, int(j0i), int(j1i)))
    return res
    
```
